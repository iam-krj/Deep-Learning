#Optimization - Gradient Descent 

```python
functionGradient = lambda x: 2*x*x
```

```python
learningRate = 0.01
precisionVar = 0.000001
stepSize = 1
maxEpoch = 10000
iterCounter= 0
currentX = 3

while stepSize > precisionVar and iterCounter < maxEpoch:
    prev_x = currentX
    currentX = currentX - learningRate * functionGradient(prev_x)
    stepSize = abs(currentX - prev_x)
    iterCounter = iterCounter+1
    print("Iteration: {0}, Value: {1}".format(iterCounter,currentX))

print("Minimum occurs at {0} & in the Iteration {1}".format(currentX,iterCounter ))
```
